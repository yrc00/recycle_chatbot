## 전국

```python
import pandas as pd

df = pd.read_excel('폐기물처리_전국.xlsx', sheet_name='Sheet1', index_col=0)

# 숫자로 변환할 열 목록 선택 (첫 번째 열은 항목명이라 제외)
cols_to_convert = df.columns[0:]

# 쉼표 제거 후 float로 변환
df = df[cols_to_convert].replace(',', '', regex=True).astype(float)

df
```

```python
# 계산: 재활용률, 물질재활용률, 에너지회수율 등
df["재활용률(%)"] = (df["재활용"] / df["발생량"]) * 100
df["물질재활용률(%)"] = (df["물질재활용"] / df["발생량"]) * 100
df["에너지회수율(%)"] = (df["에너지회수"] / df["발생량"]) * 100
df["소각률(%)"] = (df["소각"] / df["발생량"]) * 100
df["매립률(%)"] = (df["매립"] / df["발생량"]) * 100
```

```python
summary = df.describe()
summary
```

```python
recycle_summary = df[["발생량", "재활용률(%)", "물질재활용률(%)", "에너지회수율(%)", "소각률(%)", "매립률(%)"]]
recycle_summary
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns

# 예시용 palette
palette = sns.color_palette("Set2", len(df))

# 가로 막대그래프 그리기
plt.figure(figsize=(14, 6))
plt.barh(df.index, df['발생량'], color=palette)
plt.title('항목별 발생량')
plt.xlabel('발생량 (톤)')
plt.ylabel('항목')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns

# 사용할 컬럼
target = ["재활용률(%)", "물질재활용률(%)", "에너지회수율(%)", "소각률(%)", "매립률(%)"]

# palette 설정
palette = sns.color_palette("Set2", len(target))

df_rows = df.loc[df['소각률(%)'] != 0].index

for row in df_rows:
    values = df.loc[row, target]

    # 파이 차트 그리기
    plt.figure(figsize=(8, 8))
    plt.pie(values, labels=target, colors=palette, autopct="%.1f%%", startangle=140)
    plt.title(f"{row} 항목별 비율")
    plt.tight_layout()
    plt.show()
```
